```python
import pandas as pd
file_path = '../../data/games.json'
df = pd.read_json(file_path, orient='index')
df = df.reset_index()
df.rename(columns={'index': 'AppID'}, inplace=True)

# Display the first few rows of the DataFrame
print(df.head())

# df.info()
# df.head()
```

```python
dupes = df[df.duplicated(['AppID', 'name'], keep=False)]
dupes.head(10)
```

```python
df.drop_duplicates(['AppID', 'name'], keep='first', inplace=True)
dupes = df[df.duplicated(['AppID', 'name'], keep=False)]
dupes.head(5)
```
